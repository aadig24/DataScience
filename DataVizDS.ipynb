```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from ydata_profiling import ProfileReport
```

```python
data = pd.read_csv('04_Visualization_Practical_Session_Data.csv')
```

```python
data.head()
```

```python
data = data.rename(columns={"Unnamed: 0": "ID"})
data.head()
```

Using Seaborn for Data Visualization

```python
sb.set_style("white")
```

```python
#Countplot
sb.countplot(x="pclass",data=data)
```

```python
sb.countplot(y="pclass",data=data)
```

```python
#setting the color
sb.countplot(x="survived", data=data, color='green')
```

```python
#using color palettes
sb.countplot(x="sex", data=data, palette="Blues_d")
```

```python
#Reducing chartjunk by variation in color palette
sb.countplot(x="sex", data=data, palette="mako")
sb.despine()
```

```python
#Variation of color palette
sb.countplot(x="sibsp", data=data, palette="coolwarm")
```

```python
#Histogram, number of bins can also be specified
sb.histplot(x="age", data=data)
```

```python
sb.histplot(x="age", data=data, bins=5)
```

```python
#use subplots from matplotlib for integration of different plots
fig, axs = plt.subplots(2,2, figsize=(10,10))
sb.histplot(x="age", data=data, bins=2, ax=axs[0,0])
sb.histplot(x="age", data=data, bins=5, ax=axs[0,1])
sb.histplot(x="age", data=data, bins=10, ax=axs[1,0])
sb.histplot(x="age", data=data, bins=100, ax=axs[1,1])
plt.show()
```

```python
#commulated density function
sb.displot(x="age", data=data, kind="ecdf")
```

```python
# overlapping plots using FacetGrid
fg = sb.FacetGrid(data, hue="pclass", aspect=3, palette="crest")
fg.map(sb.kdeplot, "age", fill=True)
fg.set(xlim=(0, 80));
```

Visualization of Relations

```python
# 1.categorical -> categorical (ex.sex-survived) = barplot
sb.barplot(x="sex", y="survived", data=data, palette="crest", errorbar=None)
```

```python
# multiple categories via hue
sb.catplot(x="sex", y="survived", hue="pclass", kind="bar", data=data, palette="crest", errorbar=None)
```

```python
# 2.categorical -> numerical (ex. pclass-age) =
sb.swarmplot(x="pclass",y="age", data=data, palette="crest")
```

```python
# better then swarmplot (if there needs more data to be visualized)
sb.stripplot(x="pclass",y="age", data=data, palette="crest")
```

```python
sb.boxplot(x="pclass",y="age", data=data, palette="crest")
```

```python
sb.violinplot(x="pclass",y="age", data=data, palette="crest")
```

```python
# 3.numerical->numerical = linplot, regplot or scatterplots
sb.scatterplot(x="age", y="fare", data=data)
```

```python
sb.regplot(x="age", y="fare", data=data, fit_reg=False)
```

```python
sb.regplot(x="age", y="fare", data=data, fit_reg=False, marker="*")
```

```python
sb.regplot(x="age", y="fare", data=data, scatter=False)
```

```python
sb.lineplot(x="age", y="fare", data=data)
```
